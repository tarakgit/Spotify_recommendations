```python
# This text has some unwanted text like 0000,01,12,007...numeric stringvalues and 1x,12pm.combination strings , lets remove them
import re


def custom_cleaning(text):
    # Remove purely numeric words
    text = re.sub(r'\b\d+\b', '', text)
    
    # Remove time-related patterns (e.g., 12am, 3pm)
    text = re.sub(r'\b\d{1,2}[ap]m\b', '', text)
    
    # Remove other words containing numbers
    text = re.sub(r'\b\w*\d\w*\b', '', text)
    
    return text


# Apply the function to the text data
#df['text'] = df['text'].apply(remove_numeric_strings)
# Apply the function to the text data using .loc to avoid SettingWithCopyWarning
df.loc[:, 'text'] = df['text'].apply(custom_cleaning)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Sample data (replace this with your actual data)

#df = pd.DataFrame(data)

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the text data
X = vectorizer.fit_transform(df['text'])

# Extract the terms (feature names)
terms = vectorizer.get_feature_names_out()

# Convert the terms to a DataFrame
terms_df = pd.DataFrame(terms, columns=['Term'])

# Save the terms to a separate CSV file
terms_df.to_csv('terms.csv', index=False)

# Convert the DTM to a DataFrame without column names
dtm_df = pd.DataFrame(X.toarray())

# Display the Document-Term Matrix
print(dtm_df)

# Save the Document-Term Matrix to a CSV file without column names
#dtm_df.to_csv('document_term_matrix_no_column_names.csv', index=False, header=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity, pairwise_distances
from sklearn.metrics import homogeneity_score, completeness_score


# TFIDF on Document-Term Matrix to get TF-IDF scores
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(dtm_df)

# Conversion into  TF-IDF matrix to DF
tfidf_df = pd.DataFrame(tfidf_matrix.toarray())

# 'genre_labels' 
genre_labels = df['genre'].tolist()

# Split the data into training and testing sets (80% train, 20% test)
train_data, test_data, train_labels, test_labels = train_test_split(tfidf_df, genre_labels, test_size=0.2, random_state=42)


unique_categories = len(set(genre_labels))

# Perform KMeans clustering 
kmeans = KMeans(n_clusters=unique_categories, random_state=42)
kmeans.fit(train_data)


cosine_similarities = cosine_similarity(test_data, kmeans.cluster_centers_)


predicted_clusters = np.argmax(cosine_similarities, axis=1)

homogeneity = homogeneity_score(test_labels, predicted_clusters)
completeness = completeness_score(test_labels, predicted_clusters)

print(f"Homogeneity Score: {homogeneity:.4f}")
print(f"Completeness Score: {completeness:.4f}")


category_labels = {i: label for i, label in enumerate(set(genre_labels))}

# Print predicted clusters and cosine similarities for each test document
for i, (predicted_cluster, cosine_sim) in enumerate(zip(predicted_clusters, np.max(cosine_similarities, axis=1))):
    print(f"Test Document {i+1}: Predicted Category: {category_labels[predicted_cluster]}, Cosine Similarity: {cosine_sim:.4f}")


```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import homogeneity_score, completeness_score, silhouette_score


tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(dtm_df)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray())
genre_labels = df['genre'].tolist()


train_data, test_data, train_labels, test_labels = train_test_split(tfidf_df, genre_labels, test_size=0.2, random_state=42)

# LDA to reduce dimensionality
lda = LDA(n_components=min(len(set(train_labels)) - 1, train_data.shape[1]))
train_data_lda = lda.fit_transform(train_data, train_labels)
test_data_lda = lda.transform(test_data)


unique_categories = len(set(genre_labels))

# KMeans clustering on the LDA-reduced
kmeans = KMeans(n_clusters=unique_categories, random_state=42)
kmeans.fit(train_data_lda)

#  cosine similarity between each test document and cluster centroids
cosine_similarities = cosine_similarity(test_data_lda, kmeans.cluster_centers_)

# cosine similarity for each test document
predicted_clusters = np.argmax(cosine_similarities, axis=1)


category_labels = {i: label for i, label in enumerate(set(genre_labels))}
for i, (predicted_cluster, cosine_sim) in enumerate(zip(predicted_clusters, np.max(cosine_similarities, axis=1))):
    print(f"Test Document {i+1}: Predicted Category: {category_labels[predicted_cluster]}, Cosine Similarity: {cosine_sim:.4f}")

# Compute homogeneity and completeness scores
homogeneity_lda = homogeneity_score(test_labels, predicted_clusters)
completeness_lda = completeness_score(test_labels, predicted_clusters)
silhouette_avg = silhouette_score(test_data_lda, predicted_clusters)

print(f"LDA - Homogeneity Score: {homogeneity_lda}")
print(f"LDA - Completeness Score: {completeness_lda}")
print(f"LDA - Silhouette Score: {silhouette_avg}")
```

```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import homogeneity_score, completeness_score


tfidf_transformer = TfidfTransformer()


tfidf_matrix = tfidf_transformer.fit_transform(dtm_df)


tfidf_df = pd.DataFrame(tfidf_matrix.toarray())


genre_labels = df['genre'].tolist()


train_data, test_data, train_labels, test_labels = train_test_split(tfidf_df, genre_labels, test_size=0.2, random_state=42)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95)  # Retain 95% of the variance
train_data_pca = pca.fit_transform(train_data)
test_data_pca = pca.transform(test_data)


unique_categories = len(set(genre_labels))

#  KMeans clustering on the PCA-reduced data
kmeans = KMeans(n_clusters=unique_categories, random_state=42)
kmeans.fit(train_data_pca) 


cosine_similarities = cosine_similarity(test_data_pca, kmeans.cluster_centers_)


predicted_clusters = np.argmax(cosine_similarities, axis=1)


category_labels = {i: label for i, label in enumerate(set(genre_labels))}


for i, (predicted_cluster, cosine_sim) in enumerate(zip(predicted_clusters, np.max(cosine_similarities, axis=1))):
    print(f"Test Document {i+1}: Predicted Category: {category_labels[predicted_cluster]}, Cosine Similarity: {cosine_sim:.4f}")


homogeneity_pca = homogeneity_score(test_labels, predicted_clusters)
completeness_pca = completeness_score(test_labels, predicted_clusters)

print(f"PCA - Homogeneity Score: {homogeneity_pca}")
print(f"PCA - Completeness Score: {completeness_pca}")

```
